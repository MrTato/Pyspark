```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('Practice').getOrCreate()
```

```python
df_pyspark = spark.read.csv('test1.csv', header=True, inferSchema=True)
```

```python
df_pyspark.show()
```

```python
# drop the columns
df_pyspark.drop('Name').show()
```

```python
# deletes rows with null values
df_pyspark.na.drop().show()
```

```python
# drops the row if it finds one row
df_pyspark.na.drop(how='any').show()
#drops the row if all the values in the row are null
df_pyspark.na.drop(how='all').show()
```

```python
# threshold: if at most 2 null values are found, keep the row
df_pyspark.na.drop(how='any', thresh=2).show()
```

```python
# subset: drop nan values only in a specific column
df_pyspark.na.drop(how='any', subset=['Age']).show()
```

```python
# replace nan values with specified value
df_pyspark.na.fill('Missing Values', ['Age', 'Name']).show()
```

```python
df_pyspark.show()
```

```python
from pyspark.ml.feature import Imputer

imputer = Imputer(
    inputCols=['Age'],
    outputCols=["{}_imputed".format(column_name) for column_name in ['Age']]
).setStrategy('mean')
```

```python
# Add imputation cols to df
imputer.fit(df_pyspark).transform(df_pyspark).show()
```
